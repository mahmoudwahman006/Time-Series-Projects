```python
import pandas as pd

# Load the dataset
data = pd.read_csv('data.csv')

# Display the first 5 rows of the dataframe
data
```

Our dataset consists of the following columns:

    - Open Time: The timestamp at which the price bar started.
    - Open: The price at which the first trade occurred within the price bar.
    - High: The highest price at which a trade occurred within the price bar.
    - Low: The lowest price at which a trade occurred within the price bar.
    - Close: The price at which the last trade occurred within the price bar.
    - Volume: The number of shares or contracts traded in a security or an entire market during the price bar.
    - Close Time: The timestamp at which the price bar ended.
    - Quote asset volume: The volume of trading of this asset, quoted in the asset currency.
    - Number of trades: The number of trades that were executed during the price bar.
    - Taker buy base asset volume: The volume of the asset that has been bought by takers.
    - Taker buy quote asset volume: The volume of the asset that has been sold by takers.
    - Target: The specific trade is positive or negative in the next minute

The 'Open Time' and 'Close Time' are in UNIX timestamp format. For better understanding, we will convert these UNIX timestamps to a more human-readable format. Let's do that next.

```python
# Convert 'Open Time' and 'Close Time' from unix timestamp to human-readable time
data['Open Time']  = pd.to_datetime(data['Open Time'], unit='ms')
data['Close Time'] = pd.to_datetime(data['Close Time'], unit='ms')

# Display the first 5 rows of the dataframe
data.head()
```

```python
data.tail(158)
```

Now that the 'Open Time' and 'Close Time' columns are in a more human-readable format, we can better understand the data. Each row in the data represents one minute of trading activity.

The next step is to check the data for any missing values. This is an important part of the data cleaning process because missing data can lead to misleading analysis results and errors in the machine learning models. Let's check for missing values next.

```python
# Check for missing values in the dataframe
data.isnull().sum()
```

Great! There are no missing values in the dataset. This means we can move on to the next step - Feature Engineering.

In the Feature Engineering step, we try to create new features that might help improve the model's performance. In our case, we can create the following features:

    - Moving Average: This is the average price of the stock over a certain period of time, and it helps smooth out price fluctuations and highlight the overall price trend.
    - Standard Deviation (Volatility): This measures the dispersion of a set of data from its mean. In finance, standard deviation is used as an indicator of market volatility.

Let's calculate these features now.

```python
# Calculate the returns from the 'Close' price
data['Return'] = data['Close'].pct_change()

#Calculate the 7-period moving average
data['MA7'] = data['Close'].rolling(window=7).mean() # window => size of moving

# Calculate the standard deviation (volatility) over a 7-period timescale
data['Volatility'] = data['Return'].rolling(window=7).std()
```

We have now calculated the return, moving average over 7 periods (MA7), and volatility over 7 periods. These are the new features that we have created:

    - Return: The percentage change in the 'Close' price compared to the previous period.
    - MA7: The average 'Close' price over the past 7 periods.
    - Volatility: The standard deviation of the 'Return' over the past 7 periods.

Please note that for the first few rows, the MA7 and Volatility values are NaN because there isn't enough past data to calculate these values. We'll handle these NaN values later.

Now we can move on to the next step - Model Building. In this step, we will build a machine learning model that will predict whether the return in the next minute will be positive or negative.

However, before we proceed with model building, let's prepare our data by doing the following:

    - Remove the rows with NaN values (which were caused by the moving average and volatility calculations).
    - Split the data into a training set (used to train the model) and a testing set (used to evaluate the model). We'll use 70% of the data for training and 30% of the data for testing.


```python
from sklearn.model_selection import train_test_split, cross_val_score

# Drop the rows with NaN values
data = data.dropna()

# Define the feature columns
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'Quote asset volume', 'Number of trades', 
            'Taker buy base asset volume', 'Taker buy quote asset volume', 'Return', 'MA7', 'Volatility']

# Define the target column
target = 'Target'

# Split the data into features and target
X = data[features]
y = data[target]

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)
```

We have prepared our data for model building. Here's what we did:

    - Removed rows with NaN values.
    - Split the data into a training set and a testing set. The training set contains 70% of the data and the testing set contains the remaining 30%.

Now we're ready to move to the next step - Model Building.

We will use a Random Forest classifier for our trading model. Random Forest is a type of ensemble machine learning model that combines multiple smaller models to create a more robust and accurate model. It is also a good choice for this task because it can capture complex patterns in the data, and it does not require feature scaling.

Let's train our Random Forest model using the training data. We will use scikit-learn, a popular machine learning library in Python.

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest classifier
clf = RandomForestClassifier(random_state=0)

# Fit the model on the training data
clf.fit(X_train, y_train)

# Evaluate the model using cross-validation
cv_scores = cross_val_score(clf, X_train, y_train, cv=5)  # 5-fold cross-validation
```

Our Random Forest model has been trained using the training data. The model learned the patterns in the training data to predict whether the return will be positive or negative in the next minute.

Here's what each part of the code does:

    1.We import the RandomForestClassifier from scikit-learn, which allows us to create a Random Forest classifier.

    2.We create an instance of the Random Forest classifier with RandomForestClassifier(random_state=0). Setting random_state=0 ensures reproducibility.

    3.We fit the model on the training data using clf.fit(X_train, y_train), where X_train contains the features and y_train contains the target variable.

    4.Finally, we evaluate the model's performance using cross-validation. We use cross_val_score from scikit-learn to perform 5-fold cross-validation (cv=5). This function returns an array of scores, one for each fold, allowing us to assess the model's performance more accurately.




```python
# Print the cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Mean CV score:", cv_scores.mean())
```

The cross-validation scores are all 1.0, indicating perfect accuracy for each fold. The mean CV score is also 1.0, confirming the excellent performance of the model across all folds.


```python
# Predict on the test set
test_predictions = clf.predict(X_test)

# Optionally, evaluate the model on the test set
test_accuracy = clf.score(X_test, y_test)
print("Test accuracy:", test_accuracy)
```

The test accuracy is 1.0, indicating that the model achieved perfect accuracy on the unseen test data. This high accuracy suggests that the model generalizes well to new, unseen data, further validating its effectiveness for the trading task.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Make predictions on the test set
test_predictions = clf.predict(X_test)

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, test_predictions)

# Display the confusion matrix
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False,
            xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

* This confusion matrix will help you evaluate the performance of your model by showing the counts of true positive, false positive, true negative, and false negative predictions.


* The confusion matrix provides valuable insights into the model's performance by revealing how well it predicts each class. In this case, the absence of false positives and false negatives indicates that the model performs exceptionally well on the test data, further confirming its effectiveness for the trading task.

The confusion matrix is a 2x2 matrix where:

    The top-left cell represents True Negatives (TN): Actual negative cases correctly predicted as negative.
    
    The top-right cell represents False Positives (FP): Actual negative cases incorrectly predicted as positive.
    The bottom-left cell represents False Negatives (FN): Actual positive cases incorrectly predicted as negative.
    The bottom-right cell represents True Positives (TP): Actual positive cases correctly predicted as positive.

    There are 0 False Negatives (actual positive cases incorrectly predicted as negative).
    
**This indicates that the Random Forest classifier made no errors on the test set, achieving perfect predictions for both positive and negative classes.**
